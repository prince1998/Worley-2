### Haar Cascade Face & Eye Detection
Haar Cascade is a machine learning object detection algorithm used to identify objects in an image or video and based on the concept of features. They are massive XML files with a lot of feature sets and these feature sets corresponds to very specific object.

```python
import cv2
import numpy as np

face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
#capture = cv2.VideoCapture(0) #0 is used to bring in my own camera feed 
```

```python
while True:
    #ret,img = capture.read()
    img = cv2.imread('Priyansh.jpeg')
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    faces = face_cascade.detectMultiScale(gray,1.3,5) #Depending on size of image you use and likelihood that you will find that image, you will change this
    for (x,y,w,h) in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2) #Starting point: x,y; Ending point: x+w,y+h, 255,0,0: blue rectangle; line width:2
        roi_gray = gray[y:y+h,x:x+w] #Trying to find eye between starting point(y & x) and ending point(y+h & x+w) of face
        roi_color = img[y:y+h,x:x+w] #Same thing in color when we reimpose
        eyes = eye_cascade.detectMultiScale(roi_gray)
        for (ex,ey,ew,eh) in eyes: #ex - eyes x
            cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
    
    cv2.imshow('img',img)
    k = cv2.waitKey(30) & 0xff
    if k == 27: #break when pressed escape key
        break

#capture.release()
cv2.destroyAllWindows()
        
```
