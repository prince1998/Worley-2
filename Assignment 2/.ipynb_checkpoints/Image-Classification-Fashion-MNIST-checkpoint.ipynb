```python
import tensorflow as tf
from tensorflow import keras
```

```python
import cv2
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fashion_mnist = keras.datasets.fashion_mnist
```

```python
(train_images, train_labels), (test_images,test_labels) = fashion_mnist.load_data()
```

```python
test_labels
```

```python
cv2.imshow('lol',test_images[0])
```

```python
class_names = ['T-shirt/top','Trouser','Pullover','Dress','Coat',
              'Sandal','Shirt','Sneaker','Bag','Ankle Boot']
```

```python
train_images.shape
```

60000 images with each image having 28 x 28 pixels

```python
len(train_labels)
```

```python
test_images.shape
```

60000 images with each image having 28 x 28 pixels

```python
len(test_labels)
```

```python
plt.figure()
plt.imshow(train_images[0])
plt.colorbar()
plt.show()
```

The pixel values fall in range from 0 to 255. Now, we scale these values from 0 to 1

```python
#Preprocessing training and testing set
train_images = train_images/255.0
test_images = test_images/255.0
```

```python
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(train_images[i],cmap=plt.cm.binary)
    plt.xlabel(class_names[train_labels[i]])
plt.show()
```

```python
#Setting up the layer
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),#this transforms the 2d array of 28x28 pixels into a 1D Array of 784 pixels(28x28)
    keras.layers.Dense(128,activation=tf.nn.relu),#Dense layer containing 128 neurons
    keras.layers.Dense(10,activation=tf.nn.softmax)#Returns an array of 10 probability score that sums to 1; Each node contains a score that indicates the probability that the current image belongs to one of those 10 classes
])
```

```python
#Before model is ready for training, it needs few more settings - added during compile step
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

Loss function measures how accurate the model is, during the training and we want to minimize the loss function
<br>
Optimizer is how the model updates based on the data it sees and its loss function 
<br>
Metrics are used to monitor the training and testing steps; Accuracy (in our example) is the fraction of images that are correctly classified

### Training the model

```python
model.fit(train_images,train_labels,epochs=10)
```

### Evaluating the model by seeing how model performs on test dataset

```python
test_loss, test_acc = model.evaluate(test_images,test_labels)
print('Test Accuracy',test_acc)
```

Overfitting is when a machine learning model performs worse on testing data than on training data

### Model Predictions

```python
predictions = model.predict(test_images)
```

```python
predictions[0]
```

These describe confidence of the model that image corresponds to each 10 different articles of clothing

Model has highest confidence of this image for Ankle Boot or class number 9

```python
test_labels[0]
```

test_label is also 9. Therefore, prediction was correct

### Graph to see full set of 10 channels

```python
def plot_image(i,predictions_array,true_label,img):
    predictions_array, true_label, img = predictions_array[i],true_label[i],img[i]
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    
    plt.imshow(img,cmap=plt.cm.binary)
    
    predicted_label = np.argmax(predictions_array)
    if predicted_label == true_label:
        color = 'blue'
    else:
        color = 'red'
    
    plt.xlabel("{} {:2.0f}% ({})".format(class_names[predicted_label],
                                         100*np.max(predictions_array),
                                         class_names[true_label]),
                                         color = color)
    
def plot_value_array(i,predictions_array,true_label):
    predictions_array,true_label = predictions_array[i],true_label[i]
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    thisplot = plt.bar(range(10),predictions_array,color="#777777")
    plt.ylim([0,1])
    predicted_label = np.argmax(predictions_array)
    
    thisplot[predicted_label].set_color('red')
    thisplot[true_label].set_color('green')
```

```python
i = 0
plt.figure(figsize=(6,3))
plt.subplot(1,2,1)
plot_image(i,predictions,test_labels,test_images)
plt.subplot(1,2,2)
plot_value_array(i,predictions,test_labels)
plt.show()
```
